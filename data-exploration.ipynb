```python
import seaborn as sns
import matplotlib.pyplot as plt
```

In the detailed view we have seen if there are any missing values and found that many descriptions are missing . We can easily remove these columns but we need to make sure if the data is biased or skewed towards something. We need to make sure we are not removing any relevant data

```python
ax= plt.axes()  #creating the axis object
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)   #creating the heatmap where null values will be false ,coloured bar disabled to keep the plot simple and using ax axises object

plt.xlabel("Columns")
plt.ylabel("Missing")
```

Now we can clearly see the pattern here the books for which average rating is missing -num pages and ratings are also missing so maybe they are from other dataset this is a possibility but to make sure and see the relation that whether it is due to they are very old or very new that's why we have lost the description value as well we will now find out the correlation.

```python
import numpy as np

books["missing_description"]=np.where(books["description"].isna(),1,0)
books["age_of_book"]=2024-books["published_year"]
```

```python
# column_of_interest = books[["average_rating", "num_pages", "age_of_book", "missing_description"]]
#
# correlation_matrix = books[column_of_interest].corr(method="spearman")
# # pearson correlation when we are dealing with non continuous values so here mostly we have binary so using spearman
# sns.set_theme(style="white")
#
# # Create heatmap
# plt.figure(figsize=(8, 6))
# heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
# heatmap.set_title("Correlation Heatmap")
#
# # Show the plot
# plt.show()
column_of_interest = ["average_rating", "num_pages", "age_of_book", "missing_description"]

# Compute the Spearman correlation matrix
correlation_matrix = books[column_of_interest].corr(method="spearman")

# Set the Seaborn theme
sns.set_theme(style="white")

# Create heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation Heatmap")

# Show the plot
plt.show()
```

```python
books[(books['description'].isna())|
      (books['num_pages'].isna())|
      books['average_rating'].isna()|
      books['published_year'].isna()]
```

```python
books_complete=books[~(books['description'].isna())&
      ~(books['num_pages'].isna())&
     ~( books['average_rating'].isna())&
      ~(books['published_year'].isna())]
```
